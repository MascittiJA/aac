```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np


def dibujar(lista, eje, nombre, titulo):
    objects = ('Main', 'FIFO', 'RAND', 'LRU', 'LFU')
    y_pos = np.arange(len(objects))
    performance = lista

    plt.bar(y_pos, performance, align='center', alpha=0.5, color='blue')
    plt.xticks(y_pos, objects)
    plt.ylabel(eje)
    plt.title(titulo)

    plt.savefig('img/' + nombre)
    plt.show()

    
def dibujar2(lista, eje, nombre, titulo):
    objects = ('FIFO', 'RAND', 'LRU', 'LFU')
    y_pos = np.arange(len(objects))
    performance = lista

    plt.bar(y_pos, performance, align='center', alpha=0.5, color='blue')
    plt.xticks(y_pos, objects)
    plt.ylabel(eje)
    plt.title(titulo)

    plt.savefig('img/' + nombre)
    plt.show()

def dibujar3(lista, eje, nombre, titulo):
    objects = ('FIFO', 'RAND', 'LRU', 'LFU')
    y_pos = np.arange(len(objects))
    performance = lista

    plt.bar(y_pos, performance, align='center', alpha=0.5, color='blue')
    plt.xticks(y_pos, objects)
    plt.ylabel(eje)
    plt.title(titulo)

    plt.savefig('img/' + nombre)
    plt.show()
```
